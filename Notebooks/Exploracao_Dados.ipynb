```python
# Median price evolution over time
precos_mensais = dados.groupby('date')['price(MZN)'].mean().reset_index()
plt.figure(figsize=(10,4))
plt.plot(precos_mensais['date'], precos_mensais['price(MZN)'], color='green')
plt.title('Preo Mdio dos Produtos Agrcolas ao Longo do Tempo (MZN)')
plt.xlabel('Ano')
plt.ylabel('Preo Mdio (MZN)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 3 more frequent products
top_produtos = dados['commodity'].value_counts().head(3).index

plt.figure(figsize=(12,6))
for produto in top_produtos:
    dado = dados[dados['commodity'] == produto]
    dado = dado.set_index('date').sort_index()
    serie = dado['price(MZN)'].resample('ME').mean()
    plt.plot(serie, label=produto)

plt.title('Evoluo do Preo por Produto (Top 3)')
plt.xlabel('Ano')
plt.ylabel('Preo Mdio (MZN)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
